```python
# Задание-1:
# Реализуйте описаную ниже задачу, используя парадигмы ООП:
# В школе есть Классы(5А, 7Б и т.д.), в которых учатся Ученики.
# У каждого ученика есть два Родителя(мама и папа).
# Также в школе преподают Учителя. Один учитель может преподавать 
# в неограниченном кол-ве классов свой определенный предмет. 
# Т.е. Учитель Иванов может преподавать математику у 5А и 6Б,
# но больше математику не может преподавать никто другой.

# Выбранная и заполненная данными структура должна решать следующие задачи:
# 1. Получить полный список всех классов школы
# 2. Получить список всех учеников в указанном классе
#  (каждый ученик отображается в формате "Фамилия И.О.")
# 3. Получить список всех предметов указанного ученика 
#  (Ученик --> Класс --> Учителя --> Предметы)
# 4. Узнать ФИО родителей указанного ученика
# 5. Получить список всех Учителей, преподающих в указанном классе
class Subject:
    def __init__(self, purs,nB,nE):
        self.P,self.nB,self.nE = purs,nB,nE
    def get_class(self):
        return [self.nB,self.nE]
    def get_purs(self):
        return self.P
    
class Teacher:
    def __init__(self, nmNS,nP):
        self.nmNS,self.P = nmNS,nP
    def get_name(self):
        return self.nmNS
    def get_purs(self):
        return self.P
    
class Pupil:
    def __init__(self, nmNS,clNL,mnmNS,fnmNS):
        self.nmNS,self.mnmNS,self.fnmNS = nmNS,mnmNS,fnmNS
        self.clN,self.clL = int(clNL[:-2]),clNL[-1]
    def get_name(self):
        return self.nmNS
    def get_pname(self):
        return self.mnmNS,self.fnmNS
    def get_class(self):
        return '%s'%(self.clN)+' '+self.clL
#-----------------------------------------------    
PUP,TCH,SBJ = [],[],[]
SBJ.append(Subject('рус.яз', 5,9))
SBJ.append(Subject('матем', 5,11))
SBJ.append(Subject('физика', 8,11))
SBJ.append(Subject('труд', 5,9))
SBJ.append(Subject('физкультура', 5,11))

TCH.append(Teacher('Иван Петров','рус.яз'))
TCH.append(Teacher('Иван Иванов','мат'))
TCH.append(Teacher('Петр Петров','труд'))
TCH.append(Teacher('Джеки Чан','физкультура'))
TCH.append(Teacher('Лев Ландау','физика'))

PUP.append(Pupil('Альфред Нобиль',   '5 A','Марта Нобиль','Джозеф Нобиль'))
PUP.append(Pupil('Борис Беккер',     '5 А','Марта Беккер','Джозеф Беккер'))
PUP.append(Pupil('Вован Дуглас',     '5 Б','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Петр Осетр',       '6 В','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Сандро Чегемский', '7 Г','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Фирс Стасов',      '8 Б','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Захар Прилепин',   '9 Б','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Иван Кочерыжкин', '10 Б','Ольга Кочерыжкина','Петр Первый'))
PUP.append(Pupil('Андрей Новоселов','11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Бронислав Лев',   '11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Виктор Сувариков','11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Григорий Брауер', '11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Дима Егерь',      '11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Евгений Мартюш',  '11 А','Marta Nobile','Josef Nobile'))
PUP.append(Pupil('Жан-Клод Вандамм','11 А','Marta Nobile','Josef Nobile'))


#1  список всех классов в школе
LST = []; 
for pu in PUP: LST.append(pu.get_class())
print('# 1) список всех классов в школе')
print([e for e in set(LST)])
print('')

#2:  список всех учеников в указанном классе
clN,clL = '11','А'
LST = [e.get_name() for e in PUP if e.get_class()==clN+' '+clL]
print('# 2) список всех учеников в указанном классе: %s %s'%(clN,clL))
print([e for e in LST])
print('')

#3: список всех предметов указанного ученика
pu = 'Жан-Клод Вандамм'
print('# 3) список всех предметов указанного ученика: %s'%(pu))
cl = [e.get_class() for e in PUP if e.get_name()==pu]
clN = int(cl[0][:-2])
print([e.get_purs() for e in SBJ if (e.get_class()[0])<=clN and (e.get_class()[1])>=clN])
print('')

#4: ФИО родителей указанного ученика
print('# 4) ФИО родителей указанного ученика: %s'%(pu))
print([e.get_pname() for e in PUP if e.get_name()==pu])
print('')

#5: список всех Учителей, преподающих в указанном классе
clN = 5
print('# 5) список всех Учителей, преподающих в указанном классе: %s %s'%(clN,clL))
sbj = [e.get_purs()  for e in SBJ if e.get_class()[0]<=clN and e.get_class()[1]>=clN] 
print([e.get_name() for e in TCH if e.get_purs() in sbj])
```
