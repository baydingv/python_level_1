```python
# Задача-1: Написать класс для фигуры-треугольника, заданного координатами трех точек.
# Определить методы, позволяющие вычислить: площадь, высоту и периметр фигуры.
import math
class Point:
    def __init__(self,vx,vy):
        self.x, self.y = vx,vy
    def get(self):
        return [self.x, self.y]
    
class Poly:
    def __init__(self,*p):
        self.v,self.l = [],[]
        for e in p: self.v.append(p)
        n = len(self.v)
        for i in range(1,n):
            self.l.append(math.sqrt((p[i].x - p[i-1].x)**2 +(p[i].y - p[i-1].y)**2))
        self.l.append(    math.sqrt((p[0].x - p[n-1].x)**2 +(p[0].y - p[n-1].y)**2))
    def get_n(self):
        return len(self.v)
    def get_perimeter(self):
        ret = 0
        for e in self.l:
            ret += e
        return ret
    def get_sides(self):
        ret = []
        for e in self.l:
            ret.append(e)
        return ret
        
class Tre(Poly):
    def get_altitudes(self):
        h1 = 0.5*self.get_square()/self.l[0]
        h2 = 0.5*self.get_square()/self.l[1]
        h3 = 0.5*self.get_square()/self.l[2]
        return[h1,h2,h3]
    def get_square(self):
        p = 0.5*float(self.get_perimeter())
        return math.sqrt(p*(p-self.l[0])*(p-self.l[1])*(p-self.l[2])) 

# Задача-2: Написать Класс "Равнобочная трапеция", заданной координатами 4-х точек.
# Предусмотреть в классе методы:
# проверка, является ли фигура равнобочной трапецией;
# вычисления: длины сторон, периметр, площадь.    
class Tra(Poly):
    def is_tra_equihand(self):
        ret = False
        if self.get_n() != 4 : return ret
        w = self.v[0]
        [x0,y0] = w[0].get()
        [x1,y1] = w[1].get()
        [x2,y2] = w[2].get()
        [x3,y3] = w[3].get()
        cos13 = ((x1-x0)*(x3-x2)+(y1-y0)*(y3-y2))/(self.l[0]*self.l[2])
        cos24 = ((x2-x1)*(x0-x3)+(y2-y1)*(y0-y3))/(self.l[1]*self.l[3])
        if abs(cos13)>0.999 and self.l[1]==self.l[3]: ret = True
        if abs(cos24)>0.999 and self.l[0]==self.l[2]: ret = True
        return ret
    def get_square(self):
        w = self.v[0]
        t1 = Tre(w[0],w[1],w[2])
        ret = t1.get_square()
        t2 = Tre(w[0],w[3],w[2])
        ret += t2.get_square()
        return ret
        
p1 =     Point(0,0)
p2 =     Point(1,0)
p3 =     Point(1,5)
p4 =     Point(0,3)

t1 = Tre(p1,p2,p3)

print(t1.get_perimeter())
print(t1.get_square())
print(t1.get_altitudes())
print('--------')
t2 = Tra(p1,p2,p3,p4)
so = 'это - равнобокая трапеция' if t2.is_tra_equihand() else 'это не является равнобокой трапецией'; print(so)
so = 'площадь      = %s' % (t2.get_square()); print(so)
so = 'периметр     = %s' % (t2.get_perimeter()); print(so)
so = 'длина сторон = %s' % (t2.get_sides()); print(so)
```
