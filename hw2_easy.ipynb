```python
# Задача-2:
# Даны два произвольные списка.
# Удалите из первого списка элементы, присутствующие во втором списке.
import random
n1 = random.randint(0,200); s1 = []
n2 = random.randint(0, 50); s2 = [] 
for i in range(n1):
    s1.append('%02d' % random.randint(0,50))
for i in range(n2):
    s2.append('%02d' % random.randint(0,50))
print(s1)
so = 'длина первого списка: %d' % len(s1); print(so)  
print(s2)
so = 'длина второго списка: %d' % len(s2); print(so)  

n_rm = 0
for i in s1:
    if i in s2:
        while i in s1:
            s1.remove(i)
            n_rm += 1
so = 'удалено: %d, осталось: %d' % (n_rm, len(s1)); print(so)
print(s1)
```

```python
# Задача-3:
# Дан произвольный список из целых чисел.
# Получите НОВЫЙ список из элементов исходного, выполнив следующие условия:
# если элемент кратен двум, то разделить его на 4, если не кратен, то умножить на два.
n1 = 10; s1 = []; s2 =[]
for i in range(n1):
    s1.append(random.randint(-2000,2000))

for i1 in s1:
    if 2*(i1//2) == i1:
        i2 = i1 / 4
    else:
        i2 = i1 * 2
    s2.append(i2)
print(s1)
print(s2)
```
